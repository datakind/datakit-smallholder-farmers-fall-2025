```python
# Ensure pyarrow is installed in the notebook environment
conda install pyarrow 

import pandas as pd
import pyarrow as pa
import pyarrow.parquet as pq

# Define file paths
csv_file_path = 'ProducersDirect.csv'
parquet_file_path = './ProducersDirect.parquet'

# Read the CSV file into a Pandas DataFrame
df = pd.read_csv(csv_file_path)

# Remove all users from the UK (case-insensitive, handle NaNs)
mask = df['question_user_country_code'].str.contains('gb', case=False, na=False)
df = df[~mask]  # removing all users from the UK

# Convert the DataFrame to an Apache Arrow Table and write to Parquet
table = pa.Table.from_pandas(df)
pq.write_table(table, parquet_file_path)

print(f"CSV file '{csv_file_path}' converted to Parquet file '{parquet_file_path}' successfully.")

```
